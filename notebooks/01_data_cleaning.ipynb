# Phase 1: Data Cleaning

IBM Telco Customer Churn dataset — missing values, duplicates, validation, and data quality reporting.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT))

import pandas as pd
from src.config import DATA_PROCESSED, DATA_RAW, REPORTS_DIR
from src.data_cleaning import clean_data, load_raw_data, outlier_summary
```

```python
raw_df = load_raw_data(DATA_RAW)
print(f"Raw shape: {raw_df.shape}")
raw_df.head()
```

```python
cleaned_df, quality_report = clean_data(raw_df)
quality_report
```

## Cleaning decisions

- **TotalCharges**: converted to numeric; blank values for `tenure = 0` imputed with `MonthlyCharges`.
- **Duplicates**: removed on `customerID`.
- **Validation**: kept only valid `Churn` values (`Yes`/`No`) and non-negative tenure.
- **Outliers**: analyzed with IQR; retained because high spend and long tenure are valid business cases.

```python
outlier_report = outlier_summary(cleaned_df)
outlier_report
```

```python
REPORTS_DIR.mkdir(parents=True, exist_ok=True)
quality_report.to_csv(REPORTS_DIR / "data_quality_report.csv", index=False)
outlier_report.to_csv(REPORTS_DIR / "outlier_report.csv", index=False)
cleaned_df.to_csv(DATA_PROCESSED, index=False)
print("Cleaned data exported.")
```
